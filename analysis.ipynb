```python
import matplotlib.pyplot as plt
```

```python
df["category"].value_counts().plot(kind="bar")
plt.title("Number of Funds in Each Category")
plt.xlabel("Category")
plt.ylabel("Number of Funds")
plt.show()
```

```python
df["fund_house"].value_counts()
```

```python
df["fund_house"].value_counts().plot(kind="bar")
plt.title("Number of Funds by Fund House")
plt.xlabel("Fund House")
plt.ylabel("Number of Funds")
plt.show()
```

```python
df["category"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Percentage of Mutual Fund Categories")
plt.ylabel("")
plt.show()
```
